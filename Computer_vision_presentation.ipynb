### Image Classification

Another important subfield of Computer Vision is image classification which consists of only classifying whether an object is in the image or not with with a degree of confidence (usually a number between 0 and 1 which is the model output for that class). These types of models have the advantage of being computationally less expensive than object detection ones above and are of high in interest in a variety of sectors (healthcare (X-ray classification), security services (threat classification with video surveillance)...).

Below is an example of the Microsoft ResNet-50 model from HuggingFace used to classify an image as a "truck" or as "not truck-like".

## Image Captioning

Image captioning (or image-to-text) is a new field of study within Computer Vision. It consists for instance of giving a model both an image and a text query (a question) as input. The latter then generates a text on the image according to the prompt. 

```python
from image_to_caption import generate_captioning

image = 'truck_test.jpg'

question= 'What is in the image ?'

generate_captioning(image, question)
```
